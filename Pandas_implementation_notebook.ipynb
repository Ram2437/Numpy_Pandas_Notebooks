# Pandas library

```
import pandas as pd
```

## Core components of pandas

A **series** is essentially a column, and a **dataframe** is a multi-dimensional table madeup of collection of series

## Creating DataFrames from scratch

```
series = {
    'apples' : [3, 2, 5, 6],
    'oranges' : [34, 32, 55, 16]
}
```

```
series
```

```
df = pd.DataFrame(series)   # When we pass the **DataFrame** constructor to the series data we can
df
```

By default index gets created if index is not passed, which starts from zero

```
df = pd.DataFrame(data=series, index=['Jan', 'Feb', 'Mar', 'Apr'], ) # we can also create our own index values
df
```

```
df.loc['Feb': 'Apr'] # By using the loc method we can extrat the the data for the selected period in this case (Feb to Mar) 
```
